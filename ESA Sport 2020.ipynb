# Data Cleaning

```python
#Removing unused columns for this analysis
unused_column = ['dob','sofifa_id','player_url','long_name','body_type','real_face','nation_position',
'loaned_from','nation_jersey_number','player_tags','contract_valid_until','joined','team_position','team_jersey_number']
players=players.drop(unused_column,axis=1)
players.head()             
```

**After dropping some columns,the above shows the remaining colums left. We can see the position sides (lwb,cdm ,lb,st etc) have missing  data we need to clean**.

Displaying all position columns field for thorough data cleaning.

```python
#creating a variable called columns to store various positions and display the first five output .
columns = ['ls','st','rs','lw','lf','cf','rf','rw','lam','cam','ram','lm','lcm','cm','rcm','rm','lwb','ldm', 'cdm','rdm','rwb','lb','lcb','cb','rcb','rb']
players[columns].head()
```

We noticed that most numbers in the positional columns contain a "+" operator  with duplicate numbers.
we will remove all duplicate in order to get a clear/correct value for each columns.

```python
#removing all duplicate values
for col in columns:
    players[col] = players[col].str.split('+').str[0]
    
players[columns] = players[columns].fillna(0)
```

```python
#changing the dtype to 'int'
players[columns] = players[columns].astype('int')
players.info()
```

```python
players[columns]
```

```python
#replacig 0 with the mean of the column
for col in columns:
    players[col] = players[col].replace(0,(players[col].mean()))
```

 Most of the gamers only pay attention to the six main attributes **(dribbling,defending,physic,passing,shooting and pace)** in order to determine players overall ratings.

```python
#creating a variable called attritubes  to store all six player feature and dispaly the head and tail of the table.

attributes = ['dribbling','defending','physic','passing','shooting','pace']
players[attributes]
```

Since there are still missing values after viewing the table, we will find the number of missing values.

```python

#checking out all the missing value in the table
players[attributes].isna().sum()
```

We  can replace the **0 values** in the positional columns with the **mean of the column**. This can help to fill in missing values and improve the accuracy of your data. 

```python
#filling zero values with mean of the columnn
for att in attributes:
    players[att] = players[att].fillna(players[att].mean())
```
